```python
df = pd.read_csv('my_data/train.csv')
```

```python
df['Sex'] = df.Sex.replace(['male', 'female'], [0, 1])
df['Embarked'] = df.Embarked.replace(['S', 'C', 'Q'], [0, 1, 2])
```
